```python
# Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), 
# взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x=np.linspace(0, 360, 22)
k=np.linspace(0,10,22)
np.delete(k,1)
k_a=k*x
plt.plot(x,np.vectorize(lambda a:math.cos(a))(x))
plt.plot(x,np.vectorize(lambda a : math.cos(a))(k_a))
```
